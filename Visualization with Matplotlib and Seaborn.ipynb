# Controlling the size and number ofbins

# MULTIPLE HISTOGRAMS

```python
setosa_df=flowers_df[flowers_df.species=='setosa']
versicolor_df=flowers_df[flowers_df.species=='versicolor']
viginca_df=flowers_df[flowers_df.species=='virginica']
```

```python
setosa_df
```

```python
viginca_df
```

```python
versicolor_df
```

```python
import numpy as np
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.9,bins=np.arange(2,5,0.245))
plt.hist(versicolor_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.9,bins=np.arange(2,5,0.245))
plt.hist(versicolor_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25))
plt.hist(viginca_df.sepal_width,alpha=0.6,bins=np.arange(2,5,0.25))
plt.legend(['setosa','virginica','versicolor'])
plt.title('comparing distribution of sepal width')
```

```python
plt.title('distribution of sepal width')
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,viginca_df.sepal_width],bins=np.arange(2,5,0.25),stacked=True);
plt.legend(['setosa','vignicia','versicolor'])
```
